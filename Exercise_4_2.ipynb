```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(0)
```

```python
X1 = [1, 2, 3, 4, 5]
Y1 = [2.0, 2.5, 6.0, 14.5, 34.0]
plt.scatter(X1, Y1, marker="*", color="black")
```

```python
coefs: list = [4, -1, 21]

def polynomial(coefs: list, x) -> int:
    n = len(coefs)
    return sum([coefs[i]*x**i for i in range(n)])

def draw_polynomial(coefs: list) -> None:
    n = len(coefs)
    x = np.linspace(-2, 2, 1000)
    plt.ylim(0, 40)
    plt.plot(x, sum([coefs[i]*x**i for i in range(n)]), linestyle="-", color="black")

draw_polynomial(coefs)
```

```python
X = []
Y = []
for i in range(40):
    x = random.uniform(-2, 2)
    y = polynomial(coefs=coefs, x=x) + random.gauss(0, 2)
    X.append(x)
    Y.append(y)
```

```python
plt.scatter(X, Y)
```

```python
import turicreate as tc
```

```python
data = tc.SFrame({"x": X, "y": Y})
data
```

```python
for i in range(2, 5):
    col_name = f"x^{i}"
    data[col_name] = data["x"].apply(lambda x: x**i)
data
```

```python
train, test = data.random_split(.8)
```

```python
def display_result(model):
    coefs = model.coefficients
    print("Training error (rmse): ", model.evaluate(train))
    print("Tasting error (rmse): ", model.evaluate(test))
    plt.scatter(train["x"], train["y"], marker="o")
    plt.scatter(test["x"], test["y"], marker="^")
    draw_polynomial(coefs=coefs["value"])
    plt.show()
    print("Polynomial coefficients")
    print(coefs["name", "value"])
```

```python
model_L1_reg = tc.linear_regression.create(
    train, target="y", l1_penalty=0.0, l2_penalty=0.1, verbose=True, validation_set=None
)
```

```python
display_result(model_L1_reg)
```
